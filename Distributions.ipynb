```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dv = np.random.exponential(1/5, 1000)
plt.hist(dv, normed=True, bins=200,lw=0,alpha=.99, facecolor='red')
# Add a vertical line at the mean.
plt.axvline(dv.mean(), color='b', linestyle='solid', linewidth=2)
# Add a vertical line at one standard deviation above the mean.
plt.axvline(dv.mean() + dv.std(), color='b', linestyle='dashed', linewidth=2)
# Add a vertical line at one standard deviation below the mean.
plt.axvline(dv.mean()-dv.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
```
